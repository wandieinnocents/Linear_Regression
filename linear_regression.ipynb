```python
# import library
import matplotlib.pyplot as plt

# Create the arrays that represent the values of the x and y axis:
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show()
```

```python
# to draw the line
from scipy import stats
```

```python
# method that returns some important key values of Linear Regression
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
# function that uses the slopoe and intercept to return a new value
# This is where in the y-axis the corresponding x value will be placed
def myfunc(x):
  return slope * x + intercept
```

```python
# Run each value of the x array through the function
mymodel = list(map(myfunc, x))
```

```python
#Draw original scatter plot 
plt.scatter(x, y)
```

```python
# Draw the line of regression
plt.plot(x, mymodel)
plt.show() 
```

```python
# combination of the two diagrams
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show() 
```
